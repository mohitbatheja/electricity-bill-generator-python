<center><H1>Electricity Bill Management System</center>

<h3>Objective</h3>
Write a Python program that calculates the electricity bill based on units consumed, applies fixed charges, taxes, and discounts, then displays the final bill.

<H2>Problem Statement</H2><br>

Develop a Python program that accepts consumer details and calculates the final electricity bill according to the following business rules.

# Write a Python program that takes the following inputs:** <br>

• consumer_name (string) <br>
• consumer_id (int) <br>
• units_consumed (int) <br>
• senior_citizen (Yes/No) <br>

## **Then calculate the final electricity bill using the rules below:** <br>

### **1. Electricity Charges:** <br>

• ₹5 per unit if units_consumed ≤ 100 <br>
• ₹7 per unit if 100 < units_consumed ≤ 300 <br>
• ₹10 per unit if units_consumed > 300 <br>

### **2. Fixed Charge:** <br>

• Add a fixed charge of **₹200** to every bill. <br>

### **3. Senior Citizen Discount:** <br>

• If senior_citizen = **Yes**, provide **10% discount** on Electricity Charges. <br>
• Otherwise, no discount. <br>

### **4. GST:** <br>

• Apply **18% GST** on the amount after discount and fixed charge. <br>

### **5. Final Bill Calculation:** <br>

• Amount After Discount = Electricity Charges − Discount <br>

• GST = (Amount After Discount + Fixed Charge) × 18% <br>

• Final Bill = Amount After Discount + Fixed Charge + GST <br>

### **6. Final Output:** <br>

Print a complete Electricity Bill showing: <br>

• Consumer Name <br>
• Consumer ID <br>
• Units Consumed <br>
• Electricity Charges <br>
• Fixed Charge <br>
• Senior Citizen Discount <br>
• GST <br>
• Total Bill <br>

---

### **Sample Output Format**

```text
========== Electricity Bill ==========

Consumer Name        :
Consumer ID          :
Units Consumed       :
Electricity Charges  :
Fixed Charge         :
Discount             :
GST                  :
Total Bill           :

````python

consumer_name = input("Enter Consumer Name...")
consumer_id = int(input("Enter Consumer ID..."))
units = int(input("Enter Unit Consumed..."))
senior_citizen  = input("Enter Senior Citizen (Yes/No)")


electricity_charge = 0
fixed_charge = 200
discount = 0
gst = 0
#Electricity Charges

if units <=100:
    electricity_charge =units*5
elif units <=300:
    electricity_charge =units*7
else:
    electricity_charge = units*10

# Senior Citizen Discount
if senior_citizen == "Yes":
    discount = electricity_charge * 0.10
else:
    discount = 0

# Amount After Discount

amount_after_discount = electricity_charge - discount

# GST (18%)

gst = (amount_after_discount + fixed_charge)*0.18

# Final Bill

total_bill = amount_after_discount + fixed_charge + gst

# Output

print("\n===========Electricity Bill===============")

print("Consumer Name       :",consumer_name)
print("Consumer ID         :",consumer_id)
print("Unit Consumed      :", units)
print("Electricity Charge  :", electricity_charge)
print("Fixed Charge        :",fixed_charge)
print("Discount            :",discount)
print("GST                 :",gst)
print("Total Bill          :",total_bill)
````
